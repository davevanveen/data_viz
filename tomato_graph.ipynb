```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
# enter tom data for systems 1, 2
t1 = np.array([59, 52, 71, 81, 77, 58])
t2 = np.array([84, 73, 69, 88, 98, 88])

# get weeks, w switch b/w systems 1,2 at 0
w = np.arange(-len(t1), len(t2)+1, 1)
w = np.delete(w, len(t1))
w1, w2 = w[:6], w[6:]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(w1, t1, 'r:')
ax.plot(w2, t2, 'r:')
ax.set_facecolor('0.8')
ax.tick_params(axis='both', labelsize=14)
ax.tick_params(axis='y', colors='red')
ax.tick_params(axis='x', colors='white')
ax.set_ylim(bottom=45, top=105)

plt.savefig('tom_graph.png', bbox_inches='tight', dpi=600)
```
